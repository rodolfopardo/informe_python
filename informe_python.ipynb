### Para intentar predecir resultados y entender posibles resultados de nuestras acciones como empresa, se aplican regresiones lineales para predecir posibles resultados de ORDER PLACED. 
### Se probarán varios modelos y se evaluarán con los coeficientes de determinación correspondientes.

Formula

y = a + bx

## Caso 1: ¿Queremos intentar descifrar cuantas ordenes podemos obtener si aumentamos a 370.000 las visitas a Result Page en Brasil? 

```python
#Valor de X = 370.000 visitas
#Calculamos nuestra primera regresión lineal con Stats de Scipy
#Calculamos pendiente, intercepto, coeficiente r, p_value y error

slope, intercept, r_value, p_value, std_err = stats.linregress(brasil['Results_Page'], brasil['Order_Placed'])
r2 = r_value**2
```

```python
print('El valor de nuestra pendiente es: {:.2f}'.format(slope))
print('El valor de nuestro intercepto es: {:.2f}'.format(intercept))
print('El valor de nuestro r coeficiente de determinación es: {:.2f}'.format(r_value))
print('El valor de nuestro r2 es: {:.2f}'.format(r2))
print('El valor de nuestro p_value es de: {}'.format(p_value))
print('Tenemos un error de {:.2f} en nuestro modelo'.format(std_err))
```

```python
y = intercept + (slope*370000)
```

```python
print('Si tuviesemos un valor de 370.000 visitas a result_page podemos predecir que vamos a tener {0:.0f} órdenes'.format(y))
```

### Al tener un p_value menor a 0.05, rechazamos una hipótesis nula por lo que nuestra regresión lineal, a pesar de tener un r2 relativamente bajo, genera un buen parámetro de predicción.

```python
#Graficamos nuestra regresión lineal 

plt.plot(brasil['Results_Page'], brasil['Order_Placed'], 'o', label='datos originales')
plt.plot(brasil['Results_Page'], intercept + slope*brasil['Results_Page'], 'r', label='línea ajustada')
plt.legend()
plt.xlabel('Valores visitas Results_Page')
plt.ylabel('Valores visitas Order Placed')
plt.title('Regresión Lineal Results_Page - Order Placed')
plt.show()
```

## Caso 2: ¿Queremos intentar descifrar cuantas ordenes podemos obtener si aumentamos a 250.000 visitas a Profile_Page en Brasil?

```python
#Calculamos nuestra segunda regresión lineal con Stats de Scipy
#Calculamos pendiente, intercepto, coeficiente r, p_value y error


slope, intercept, r_value, p_value, std_err = stats.linregress(brasil['Profile_Page'], brasil['Order_Placed'])
r2 = r_value**2
```

```python
print('El valor de nuestra pendiente es: {:.2f}'.format(slope))
print('El valor de nuestro intercepto es: {:.2f}'.format(intercept))
print('El valor de nuestro r coeficiente de determinación es: {:.2f}'.format(r_value))
print('El valor de nuestro r2 es: {:.2f}'.format(r2))
print('El valor de nuestro p_value es de: {}'.format(p_value))
print('Tenemos un error de {:.2f} en nuestro modelo'.format(std_err))
```

```python
y = intercept + (slope*250000)
```

```python
print(y)
```

```python
print('Si tuviesemos un valor de 250.000 visitas a profile_page podemos predecir que vamos a tener {0:.0f} órdenes'.format(y))
```

## Caso 3: ¿Queremos intentar descifrar cuantas ordenes podemos obtener si aumentamos a 80.000 visitas la página de Checkout en Brasil?

```python
print('El valor de nuestra pendiente es: {:.2f}'.format(slope))
print('El valor de nuestro intercepto es: {:.2f}'.format(intercept))
print('El valor de nuestro r coeficiente de determinación es: {:.2f}'.format(r_value))
print('El valor de nuestro r2 es: {:.2f}'.format(r2))
print('El valor de nuestro p_value es de: {}'.format(p_value))
print('Tenemos un error de {:.2f} en nuestro modelo'.format(std_err))
```

```python
y = intercept + (slope*80000)
```
